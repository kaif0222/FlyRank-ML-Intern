```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/kaif0222/FlyRank-ML-Intern/main/data/raw/content_refresh_anonymized.csv")

# First First insight
print(df.columns)
print("Does content length havae any affect on engagment rate or scroll rate")
print(df['char_count'].corr(df['engagement_rate']))
print(df['char_count'].corr(df['scroll_rate']))

# Second Insight
print("Is more seach volume cause more impression?")
print("Correlation of search volume and impression", df['search_volume'].corr(df['impressions_last_30d']))


# Third Insight
print("Relation ship between seaarch volume and competition")
print("Correlation of search volume and competition", df['search_volume'].corr(df['competition']))

```
